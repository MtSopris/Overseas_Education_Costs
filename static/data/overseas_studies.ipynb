```python
import pandas as pd
import numpy as np
from glob import glob, iglob
import pymongo
import json
import os
```

```python
#read in avrage Tuition data
excel_path=os.path.join("../data",'clean_tuition.xlsx')
excel=pd.read_excel(excel_path)
tuition_df=pd.DataFrame(excel)
tuition_df.drop(tuition_df.columns[tuition_df.columns.str.contains('unnamed',case = False)],axis = 1, inplace = True)
tuition_df
```

```python
#read in level of outbound students per education level by percentage
excel_path=os.path.join("../data",'clean_student_%_by_edLevel.xlsx')
excel=pd.read_excel(excel_path)
USabroad_demogrph_df=pd.DataFrame(excel)
USabroad_demogrph_df
```

```python
#read in google country locations
csv_path=os.path.join("../data",'country_lat_lng.csv')
csv=pd.read_csv(csv_path)
lat_log_df=pd.DataFrame(csv)
lat_log_df.head(25)
```

```python
#Merge lat long to tuition_df
tuition_loc_df = pd.merge(tuition_df, lat_log_df, on="Country", how="left")
tuition_loc_df
```

```python
# drop na values from lat log column
clean_tuit_loc_df=tuition_loc_df.dropna(subset=['latitude', 'longitude'])#-test
clean_tuit_loc_df
```

```python
clean_tuit_loc_df.columns
```

```python
#filter tuition for us student cost per counrty
fgn_tuition_df= clean_tuit_loc_df[['Country','pub_fgn_bach', 'pub_fgn_mast', 'pub_fgn_phd','pvt_fgn_bach', 'pvt_fgn_mast',
       'pvt_fgn_phd', 'Abv', 'latitude', 'longitude']]
fgn_tuition_df
```

```python
#read in university ranking from 2014-2016
csv_path=os.path.join("../data",'timesData.csv')
csv=pd.read_csv(csv_path)
rank_df=pd.DataFrame(csv)
rank_df
```

```python
#get avrage ranking of uni's listed in the top 200
rank_mean_df=rank_df.groupby(['Country']).mean().reset_index()
rank_mean_df
```

```python
#add loc to ranking df
rank_loc_df = pd.merge(rank_mean_df, lat_log_df, on="Country", how="left")
rank_loc_df
```

```python
#drop countries that don't match a loc
clean_rank_loc_df=rank_loc_df.dropna(subset=['latitude', 'longitude'])#-test
clean_rank_loc_df
```

```python
#read in outbound student data
excel_path=os.path.join("../data",'clean_outbound.xlsx')
excel=pd.read_excel(excel_path)
USabroadregion_df=pd.DataFrame(excel)
USabroadregion_df.drop(USabroadregion_df.columns[USabroadregion_df.columns.str.contains('unnamed',case = False)],axis = 1, inplace = True)
USabroadregion_df
```

```python
# rename column to add loc
USabroadregion_df.rename(columns={'Destinations':'Country'}, inplace=True)
USabroadregion_df
```

```python
#add availible loc to countries for outbound students
outbound_loc_df = pd.merge(USabroadregion_df, lat_log_df, on="Country", how="left")
clean_outbound_loc_df=outbound_loc_df.dropna(subset=['latitude', 'longitude'])#-test
clean_outbound_loc_df
```

```python
#read in inbound student data
excel_path=os.path.join("../data",'clean_inbound.xlsx')
excel=pd.read_excel(excel_path)
inbound_abroad_df=pd.DataFrame(excel)
inbound_abroad_df.drop(inbound_abroad_df.columns[inbound_abroad_df.columns.str.contains('unnamed',case = False)],axis = 1, inplace = True)
inbound_abroad_df
```

```python
#rename column to add loc 
inbound_abroad_df.rename(columns={'Place of Origin':'Country'}, inplace=True)
inbound_abroad_df
```

```python
#add availible loc to countries for inbound students
inbound_loc_df = pd.merge(inbound_abroad_df, lat_log_df, on="Country", how="left")
clean_inbound_loc_df=inbound_loc_df.dropna(subset=['latitude', 'longitude']).reset_index()
clean_inbound_loc_df
inbound_df=clean_inbound_loc_df[['Country','in14/15','in15/16','in16/17','in17/18','in18/19','in19/20']]
inbound_df
```

```python
#read in avrage cost of living
csv_path=os.path.join("../data",'cost_of_living.csv')
csv=pd.read_csv(csv_path)
cost_of_living_df=pd.DataFrame(csv)
cost_of_living_df
```

```python
#filter cost of living for yearly cost
liv_cost_df=cost_of_living_df[['Country','Yearly']]
liv_cost_df
```

```python
#add cost of living to fgn student cost
fgn_tuition_df
totals_df = pd.merge(fgn_tuition_df, liv_cost_df, on="Country", how="left")
total_cost_df=totals_df.dropna(subset=['Yearly'])#.reset_index()
cost_df= total_cost_df.drop(columns=['latitude','longitude','Abv'])
cost_df
```

```python
# fgn_tuition_df
stage_df =clean_outbound_loc_df.merge(cost_df, how='outer', left_on='Country', right_on='Country')
# stage_df=totals_df.drop(subset=['Yearly']).reset_index()
stage_df
```

```python
fin_std_df =stage_df.merge(inbound_df, how='outer', left_on='Country', right_on='Country')
# stage_df=totals_df.drop(subset=['Yearly']).reset_index()
fin_std_df
```

```python
fin_std_df.to_csv('../data/final_df.csv')
```
